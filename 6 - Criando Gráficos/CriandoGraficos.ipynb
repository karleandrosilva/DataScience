```python
import pandas as pd
import matplotlib.pyplot as plt

# Criando uma Tabela
tabela = {'Título': ['Roma', 'The Irishman', 'Marriage Story', 'Okja', 'The Trial of the Chicago 7', 'Beasts of No Nation', 'Uncut Gems', 'Da 5 Bloods', 'Mudbound', 'The Social Dilemma'],
    'Ano de Lançamento': [2018, 2019, 2019, 2017, 2020, 2015, 2019, 2020, 2017, 2020],
    'Avaliação': [9.6, 9.5, 9.4, 8.6, 8.9, 9.2, 9.1, 9.2, 9.7, 8.5],
    'Número de Votos' : ['31.180', '2.268.288', '1.994.599', '20.595', '44.074', '87.801', '268.288', '2.994.599', '40.999', '144.074'],
    'Duração': [135, 209, 137, 120, 130, 137, 135, 156, 134, 94],
    'Gênero': ['Drama', 'Drama', 'Romance', 'Aventura', 'Drama', 'Drama', 'Drama', 'Drama', 'Drama', 'Documentário'],
    'País de Produção': ['México', 'EUA', 'EUA', 'Coreia do Sul', 'EUA', 'EUA', 'EUA', 'EUA', 'EUA', 'EUA']
}

# DataFrame
df = pd.DataFrame(tabela)
```

```python
df.head()
```

```python
# Analisar qual maior genero de filmes

eixo_x = df['Gênero'].value_counts().keys() # Para pegar o eixo X para o Gráfico (chave)

eixo_y = df['Gênero'].value_counts().values # Para pegar o eixo Y para o Gráfico (valores)

plt.bar(eixo_x, eixo_y) # Para exibir o Gráfico de barras

plt.title('Filmes por Gênero') # Cria o Título

#plt.xticks(rotation=45) # Rotaciona as marcaçôes do eixo X

plt.xlabel('Gêneros de Filmes') # Nome para o eixo X
plt.ylabel('Quantidades de Filmes') # Nome para o eixo Y

plt.show()
```

```python
# Analisar os melhores filmes por ano de lançamento

eixo_x = df['Ano de Lançamento'].value_counts().sort_index().keys()  # .sort_index() ->organiza em ordem decrescente

eixo_y = df['Ano de Lançamento'].value_counts().sort_index().values 

plt.plot(eixo_x, eixo_y)

plt.title('Melhores Filmes por Ano de Lançamento')

plt.xlabel('Ano de Lançamento')
plt.ylabel('Quantidades de Filmes') 

plt.show()
```
