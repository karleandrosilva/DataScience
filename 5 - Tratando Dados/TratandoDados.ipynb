```python
import pandas as pd
import numpy as np

# CRIANDO UMA TABELA ALEATÓRIA

# Definir o número de linhas e colunas para o DataFrame
num_rows = 100
columns = ['ID', 'Nome', 'Idade', 'Cidade', 'Salario']

# Gerar dados aleatórios
np.random.seed(0)  # Para reprodutibilidade

data = {
    'ID': range(1, num_rows + 1),
    'Nome': np.random.choice(['Alice', 'Bob', 'Charlie', 'David', 'Eve'], num_rows),
    'Idade': np.random.randint(20, 60, num_rows),
    'Cidade': np.random.choice(['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Porto Alegre'], num_rows),
    'Salario': np.random.randint(2000, 10000, num_rows).astype(str)
}

# Adicionar letras indesejadas em alguns valores de Salario
for i in range(0, num_rows, 10):
    data['Salario'][i] = data['Salario'][i] + '\n /Ano'

# Criar o DataFrame
df = pd.DataFrame(data)

# Salvar o DataFrame como um arquivo Excel
df.to_excel('sao_paulo.xlsx', index=False)

# Ler o arquivo Excel que acabamos de criar
df_lido = pd.read_excel('sao_paulo.xlsx')

df
```

```python
auxiliar = df.loc[0]

auxiliar
```

```python
# Função para Tratamento/Limpeza (Corrigir o preço)

def limpa_salario(linha):
    linha['Salario'] = linha['Salario'].replace('\n','')
    linha['Salario'] = linha['Salario'].replace('/','')
    linha['Salario'] = linha['Salario'].replace(' ','')
    linha['Salario'] = linha['Salario'].replace('Ano','')
    return linha # devolve já tratada
```

```python
limpa_salario(auxiliar)
```

```python
#Tratando na tabela toda

df.apply(lambda x: limpa_salario(x), axis=1)
```
